```python
import cv2
import numpy as np

#read the image 
img = cv2.imread("Images/cameraman.png")

#convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#negative
negative = 255 - gray

#log Transform 
r = gray.astype(np.float32) / 255.0
c = 1.0
log_transformed = c * np.log(1 + r)

log_transformed = cv2.normalize(log_transformed, None, 0, 255, cv2.NORM_MINMAX)
log_transformed = log_transformed.astype(np.uint8)

#power Law /gamma
gamma = 0.3
power_law = np.power(r, gamma)
power_law = (power_law * 255).astype(np.uint8)

#display results
cv2.imshow("Gray", gray)
cv2.imshow("Negative", negative)
cv2.imshow("Log transform", log_transformed)
cv2.imshow("Power law (Gamma)", power_law)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
