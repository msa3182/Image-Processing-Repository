```python
import cv2
import numpy as np
import os

img_path = "Images/Grey-scale-Lena-image-256256.png"


img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)


h, w = img.shape  


scale = 1.5
resized = cv2.resize(img, None, fx=scale, fy=scale, interpolation=cv2.INTER_LINEAR)

angle = 120
center = (w // 2, h // 2)
M_rot = cv2.getRotationMatrix2D(center, angle, 1.0)
rotated_120 = cv2.warpAffine(img, M_rot, (w, h), flags=cv2.INTER_LINEAR, borderMode=cv2.BORDER_REFLECT)

shx = 0.3
M_shear = np.float32([[1, shx, 0],
                      [0, 1,   0]])
new_w = int(w + abs(shx) * h)
sheared = cv2.warpAffine(img, M_shear, (new_w, h), flags=cv2.INTER_LINEAR, borderMode=cv2.BORDER_REFLECT)


negative = 255 - img

img_f = img.astype(np.float32)
c = 255.0 / np.log(1.0 + np.max(img_f))
log_img = (c * np.log(1.0 + img_f))
log_img = np.clip(log_img, 0, 255).astype(np.uint8)

gamma = 0.5
norm = img_f / 255.0
gamma_img = (255.0 * (norm ** gamma))
gamma_img = np.clip(gamma_img, 0, 255).astype(np.uint8)


cv2.imshow("Original (Gray)", img)
cv2.imshow("Resized", resized)
cv2.imshow("Rotated 120", rotated_120)
cv2.imshow("Sheared", sheared)
cv2.imshow("Negative", negative)
cv2.imshow("Log", log_img)
cv2.imshow("Gamma", gamma_img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
