```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
image = cv2.imread(r"Lab5-Images\parrot.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
#smoothing
smoothed = cv2.blur(image, (7, 7))

fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(image)
ax[0].set_title("Original")
ax[0].axis("off")

ax[1].imshow(smoothed)
ax[1].set_title("Smoothed 7x7")
ax[1].axis("off")
plt.show()
```

```python
#guassan filters 
gaussian5 = cv2.GaussianBlur(image, (5, 5), 0)
gaussian21 = cv2.GaussianBlur(image, (21, 21), 0)


fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].imshow(image)
ax[0].set_title("Original")
ax[0].axis("off")

ax[1].imshow(gaussian5)
ax[1].set_title("Gaussian 5x5")
ax[1].axis("off")

ax[2].imshow(gaussian21)
ax[2].set_title("Gaussian 21x21")
ax[2].axis("off")
plt.show()
```

```python
#Convert to float32
image_float = image.astype(np.float32) / 255.0

#Laplacian filter
laplacian = cv2.Laplacian(image_float, cv2.CV_32F, ksize=3)

#Sharpen
sharpened = np.clip(image_float - laplacian, 0, 1)

#Display
laplacian_display = np.clip(np.abs(laplacian), 0, 1)


fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].imshow(image_float)
ax[0].set_title("Original Image")
ax[0].axis("off")

ax[1].imshow(laplacian_display)
ax[1].set_title("Laplacian Filter Result")
ax[1].axis("off")

ax[2].imshow(sharpened)
ax[2].set_title("Sharpened Image")
ax[2].axis("off")

plt.show()
```
