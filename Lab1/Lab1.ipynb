# NumPy Exercises 

Now that we've learned about NumPy let's test your knowledge. We'll start off with a few simple tasks, and then you'll be asked some more complicated questions.

#### Import NumPy as np

````python
import numpy as np
````

#### Create an array of 10 zeros 

````python
Array = np.zeros(10)
print(Array)
````

#### Create an array of 10 ones

````python
Array = np.ones(10)
print(Array)
````

#### Create an array of 10 fives

````python
Array = np.ones(10)*5
print(Array)
````

#### Create an array of the integers from 10 to 50

````python
Array = np.arange(10,51)
print(Array)
````

#### Create an array of all the even integers from 10 to 50

````python
Array = np.arange(10,52,2)
print(Array)
````

#### Create a 3x3 matrix with values ranging from 0 to 8

````python
matrix = np.arange(9).reshape(3,3)
print(matrix)
````

#### Create a 3x3 identity matrix

````python
Identity = np.identity(3)
print(Identity)
````

#### Use NumPy to generate a random number between 0 and 1

````python
x = np.random.rand()
print(x)
````

#### Use NumPy to generate an array of 25 random numbers sampled from a standard normal distribution

````python
arr1 = np.random.randn(25).reshape(5,5)
print(arr1)
````

#### Create an array of 20 linearly spaced points between 0 and 1:

````python
Array = np.linspace(0, 1, 20)
print(Array)
````

### Numpy Indexing and Selection

Now you will be given a few matrices, and be asked to replicate the resulting matrix outputs:

````python
arr_2d = np.array([
    [5, 10, 15],
    [20, 25, 30],
    [35, 40, 45]
])

arr_2d
````

Retrieve the value 25 from the array using indexing.

````python
print(arr_2d[1,1])
````

Extract the second row of the array.

````python
print(arr_2d[1])
````

Extract the third column of the array.

````python
print(arr_2d[2])
````

Extract the bottom-right 2Ã—2 sub-array from arr_2d.

````python
sub = arr_2d[1:3, 1:3]
print(sub)
````
